Площадь прямоугольника равна 9 см. кв., а величина одного из углов равна 120 градусов. Найти стороны
прямоугольника.

![Task-8-3](Task-8-3.png)


```python
from sympy import Eq
from sympy import solve
from sympy import symbols
from sympy import pi
from sympy import sin

x,y,d = symbols("x y d")
eq1 = Eq(d ** 2 / 2 * sin(120 * pi / 180), 9)
answ1 = solve(eq1)
sys_eq = [Eq(x ** 2 + y ** 2, answ1[1] ** 2), Eq(x*y, 9)]
print(answ1)
answ2 = solve(sys_eq,[x,y])
print(answ2)
```
